```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

```python
df.shape
```

```python
df['education'].unique()
```

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
df['salary'].unique()
```

```python
100 * lower_education[lower_education['salary'] == '>50K'].shape[0] / lower_education.shape[0]
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
num_min_workers = df[(df['hours-per-week'] == min_work_hours)]
num_min_workers.shape
```

```python
num_min_workers[num_min_workers['salary'] == '>50K'].shape[0]
```

```python
100 * num_min_workers[num_min_workers['salary'] == '>50K'].shape[0] / num_min_workers.shape[0]
```

```python
df.groupby('native-country')['salary'].apply(lambda x: x == '>50K')
```
